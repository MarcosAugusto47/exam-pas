```python
import re
from pypdf import PdfReader
import pandas as pd
```

```python
reader = PdfReader("../data/raw/2020_2022/Ed_31_PAS_3_2020_2022_Conv_RA_1ª_Chamada.pdf")
number_of_pages = len(reader.pages)
```

```python
text_list = []
for i in range(number_of_pages):
    text_list.append(reader.pages[i].extract_text())
```

```python
text = ''.join(text_list)
```

```python
text
```

```python
#text = re.sub(r"[0-9]* \n.*Nome.*turno", "", text)
text = re.sub(r"N.*\n.*Nome.*subsistema", "", text)
```

```python
text = re.sub(r"[0-9]{1}\.[0-9]{1}\.[0-9]+.*\n", "", text)
```

```python
text = re.sub(r"[0-9]+ (\n )", r"\n", text)
```

```python
# remove double white spaces only if they are surrounded by letters in a string
text = re.sub(r'(?<=[a-zA-Z]) {2}(?=[a-zA-Z])', ' ', text)
```

```python
text
```

```python
text = re.sub(r"\n\* Cursos que exigem.*", "", text, flags=re.DOTALL)
```

```python
text
```

```python
# split by course
#pattern = r'  \n(?!Ciências Ambientais)'
pattern = r'(?=\n[A-Z]{1}.*  \n)'
pattern = r'(?=\n[A-Z]{1}.*\s)'
text_split = re.split(pattern=pattern, string=text)
#sections = [section for section in text_split if section.strip()]
```

```python
len(text_split)
```

```python
text_split = [text for text in text_split if re.search('[0-9]{8}', text)]
text_split = [re.sub(r"\n", "", text) for text in text_split]
```

```python
def get_approvals_by_course(course_approvals: str):
    pattern = r'(?=[0-9]{8})'
    course_segment = re.split(pattern=pattern, string=course_approvals)
    course_segment = [x.strip() for x in course_segment]

    df_course = pd.Series(course_segment[1:]).str.split('  ', expand=True)
    df_course['course'] = course_segment[0]

    return df_course
       
```

```python
def break_list_students_info(student: str):
    
    pattern = "^(.*?)\d{8} "
    result = re.search(pattern, student)
    extracted_text = result.group(0)

    student_identification = re.split('([0-9]{8})', extracted_text.strip())
    my_list = list(filter(None, student_identification))

    course_info = re.sub(re.escape(extracted_text), "", student)
    course_info = re.split(pattern=' / ', string=course_info, maxsplit=1)
    campus = course_info[0]
    course_period = course_info[1].rsplit("/", 1)
    
    my_list.append(campus)
    my_list.extend(course_period)
    
    return pd.DataFrame([my_list])
```
